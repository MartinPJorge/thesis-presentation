```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('~/Documentos/5growth/5growth-scaling/cleaned-sorted-torino.csv')
```

```python
#Selecting 'Corso Orbassano(TO)' as it is the most frequently measured
ROADNAME = 'Corso Orbassano(TO)'
LAT = 45.03639
LNG = 7.62665
LCD1 = 40097
OFFSET = 186
```
